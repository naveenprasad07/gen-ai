### HuggingFace X LangChain 

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
from langchain_huggingface import HuggingFaceEndpoint, ChatHuggingFace
from langchain_core.messages import HumanMessage, SystemMessage

# Step 1: Create the endpoint with provider="auto"
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Meta-Llama-3-8B-Instruct",
    task="text-generation",
    max_new_tokens=150,
    temperature=0.7,
    do_sample=True,
    provider="auto", 
    huggingfacehub_api_token=os.getenv("HF_TOKEN"),
)


chat = ChatHuggingFace(llm=llm, verbose=True)

messages = [
    SystemMessage(content="You are a helpful assistant."),
    HumanMessage(content="What is machine learning?"),
]

response = chat.invoke(messages)
print(response.content)
```

```python
from langchain_classic import PromptTemplate,LLMChain
template = """
Question : {question}
"""
prompt = PromptTemplate(template=template,input_variable=["question"])
print(prompt)
```

```python
llm_chain = LLMChain(llm = llm, prompt = prompt)
chat.invoke("What is machine learning?")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings  

model_name = "BAAI/bge-large-en-v1.5"  

model_kwargs = {"device": "cpu"}
encode_kwargs = {"normalize_embeddings": True}

hf = HuggingFaceEmbeddings(
    model_name=model_name,     
    model_kwargs=model_kwargs,
    encode_kwargs=encode_kwargs,
)

# Test it
texts = ["What is machine learning?", "How does AI work?"]
embeddings = hf.embed_documents(texts)
print(f"Embedding dimension: {len(embeddings[0])}")

# For a single query
query_embedding = hf.embed_query("What is machine learning?")
print(f"Query embedding dimension: {len(query_embedding)}")
```
