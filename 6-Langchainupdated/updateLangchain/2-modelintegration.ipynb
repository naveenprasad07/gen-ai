### Streaming


```python
model = init_chat_model("groq:llama-3.1-8b-instant")
```

```python
model.invoke("Write me a 200 words paragraph on Stars?")
```

```python
for chunk in  model.stream("Write me a 200 word paragraph on stars"):
    print(chunk.text,end="|",flush=True)
```

```python
for chunk in  model.stream("Write me a 200 word paragraph on stars"):
    print(chunk.content,end='|',flush=True)
```

### Batch

```python
responses = model.batch([
    "When was earth formed?",
    "When will earth die?",
    "what is fourth dimension?",
    max_concurrent = 5 ### Limit to a 5 concurrent requests
])

for response in responses:
    print(response)
```
